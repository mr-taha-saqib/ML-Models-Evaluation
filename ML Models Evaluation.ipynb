```python
import numpy as np

class NearestNeighborClassifierManual:
    def __init__(self):
        self.X_train = None
        self.y_train = None

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for test_example in X_test:
            distances = []
            for i in range(len(self.X_train)):
                distance = np.sqrt(np.sum((test_example - self.X_train[i]) ** 2))
                distances.append(distance)
            nearest_neighbor_index = np.argmin(distances)
            predictions.append(self.y_train[nearest_neighbor_index])
        return np.array(predictions)
```

```python
import numpy as np

class GaussianNaiveBayesClassifierManual:
    def __init__(self):
        self.class_priors = {}
        self.class_means = {}
        self.class_variances = {}

    def fit(self, X_train, y_train):
        for c in np.unique(y_train):
            self.class_priors[c] = np.mean(y_train == c)
            self.class_means[c] = np.mean(X_train[y_train == c], axis=0)
            self.class_variances[c] = np.var(X_train[y_train == c], axis=0)

    def predict(self, X_test):
        return np.array([np.argmax([self.class_priors[c] * np.prod(normpdf(X_test[i], self.class_means[c], np.sqrt(self.class_variances[c]))) for c in self.class_priors]) for i in range(X_test.shape[0])])
```

```python
class SupportVectorMachineClassifierManual:
    def __init__(self, learning_rate=0.001, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X_train, y_train):
        self.weights = np.zeros(X_train.shape[1])
        self.bias = 0
        for _ in range(self.epochs):
            for x, y in zip(X_train, y_train):
                margin = 1 - y * (np.dot(x, self.weights) + self.bias)
                if margin > 0:
                    self.weights += self.learning_rate * y * x
                    self.bias += self.learning_rate * y

    def predict(self, X_test):
        return np.sign(np.dot(X_test, self.weights) + self.bias)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

class ConfusionMatrix:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred
        self.n_classes = len(np.unique(y_true))
        self.matrix = self._compute_confusion_matrix()

    def _compute_confusion_matrix(self):
        matrix = np.zeros((self.n_classes, self.n_classes), dtype=int)
        for i in range(len(self.y_true)):
            matrix[self.y_true[i], self.y_pred[i]] += 1
        return matrix

    def plot(self):
        plt.figure(figsize=(8, 6))
        sns.heatmap(self.matrix, annot=True, cmap='Blues', fmt='d', xticklabels=np.arange(self.n_classes), yticklabels=np.arange(self.n_classes))
        plt.xlabel('Predicted labels')
        plt.ylabel('True labels')
        plt.title('Confusion Matrix')
        plt.show()
```

```python
class EvaluationMetrics:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred
        self.confusion_matrix = ConfusionMatrix(y_true, y_pred)
        self.metrics = self._compute_metrics()

    def _compute_metrics(self):
        tp = np.diag(self.confusion_matrix.matrix).astype(float)
        fp = np.sum(self.confusion_matrix.matrix, axis=0).astype(float) - tp
        fn = np.sum(self.confusion_matrix.matrix, axis=1).astype(float) - tp
        tn = np.sum(self.confusion_matrix.matrix).astype(float) - tp - fp - fn

        sensitivity = tp / (tp + fn)
        specificity = tn / (tn + fp)
        fpr = fp / (fp + tn)
        fnr = fn / (fn + tp)
        precision = tp / (tp + fp)

        return {
            'Sensitivity': sensitivity,
            'Specificity': specificity,
            'FPR': fpr,
            'FNR': fnr,
            'Precision': precision,
            'Recall': tp / (tp + fn),
            'F1 Score': 2 * (precision * tp / (tp + fp)) / (precision + tp / (tp + fn))
        }
```
